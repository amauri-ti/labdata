### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

# Criar um array com 10 números inteiros aleatórios entre 1 e 50
arr = np.random.randint(1, 51, size=10)

# Calcular a média e o desvio padrão do array
mean_value = np.mean(arr)
std_value = np.std(arr)

# Exibir os resultados
print("Array: ", arr)
print("Média: ", mean_value)
print("Desvio padrão: ", std_value)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
import numpy as np

# Criar um array com 10 números aleatórios
arr = np.random.rand(10)

# Calcular a raiz quadrada dos valores menores ou iguais a 0,5
arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])

print(arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
import numpy as np

# Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios
matriz_3d = np.random.rand(3, 3, 3)

# Exibir o valor máximo de cada linha de cada matriz 2D dentro da matriz 3D
max_values = np.amax(matriz_3d, axis=1)

# Exibir os resultados
print("Matriz 3D:\n", matriz_3d)
print("Valores máximos de cada linha de cada matriz 2D:\n", max_values)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
import numpy as np

# Criar uma matriz de tamanho 5x5 com valores aleatórios
matriz = np.random.rand(5, 5)

# Subtrair a média de cada linha da matriz
# axis=1 --> a linha
# axis=0 --> a coluna
media = np.mean(matriz, axis=1)
matriz_linha_media = matriz - media[:, np.newaxis]

print("Matriz original:\n", matriz)
print("Média de cada linha:\n", media)
print("Matriz com a média de cada linha subtraída:\n", matriz_linha_media)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
import numpy as np

# Criar uma matriz de tamanho 4x4 com valores aleatórios
matriz = np.random.rand(4, 4)

# Calcular a média e o desvio padrão das diagonais da matriz
diag_principal = np.diag(matriz)
diag_secundaria = np.diag(np.fliplr(matriz))
media_diag = np.mean([diag_principal, diag_secundaria])
std_diag = np.std([diag_principal, diag_secundaria])

# Exibir os resultados
print("Matriz:\n", matriz)
print("Diagonal principal:\n", diag_principal)
print("Diagonal secundária:\n", diag_secundaria)
print("Média das diagonais:", media_diag)
print("Desvio padrão das diagonais:", std_diag)
```
