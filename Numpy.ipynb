### NUMPY

```
import numpy as np

arr = np.random.rand(10)

min_value = np.min(arr)
max_value = np.max(arr)

print("Array: ", arr)
print("Valor mínimo: ", min_value)
print("Valor máximo: ", max_value)
```

```
import numpy as np

arr = np.random.rand(10)

# Substituir os valores maiores que 0,5 por 1 e os valores menores ou iguais a 0,5 por 0
arr[arr > 0.5] = 1
arr[arr <= 0.5] = 0

print(arr)
```

```
import numpy as np

matriz = np.random.rand(3, 3)

# Calcular a matriz transposta
matriz_transposta = np.transpose(matriz)


print("Matriz original:\n", matriz)
print("Matriz transposta:\n", matriz_transposta)
```

```
import numpy as np

arr = np.random.rand(10)

# Calcular a média móvel com janela de tamanho 3
janela = 3
media_movel = np.convolve(arr, np.ones(janela) / janela, mode='valid')

print("Array original:", arr)
print("Média móvel com janela de tamanho", janela, ":", media_movel)
```
