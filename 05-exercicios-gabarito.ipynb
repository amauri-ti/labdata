### 1. Some todos os números de 1 até 100

```python
soma = 0
for numero in range(1, 101):
    soma += numero

print(soma)
```

### 2. Some todos os números pares de 1 até 100

```python
soma_pares = 0

for numero in range(1, 101):
    if numero % 2 == 0:
        soma_pares += numero

print(soma_pares)
```

### 3. Some todos os números ímpares de 1 até 100

```python
soma_impares = 0

for numero in range(1, 101):
    if numero % 2 != 0:
        soma_impares += numero

print(soma_impares)
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
lista
```

```python
soma_positivos = 0

for elemento in lista:
    if (isinstance(elemento, int)) and (elemento > 0):
        soma_positivos += elemento
        
print(soma_positivos)
```
