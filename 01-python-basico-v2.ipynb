```python
a_invertido = a[::-1]
print(a_invertido)
```

```python
palavras = a.split() #dividi a frase em palavras
palavras_invertidas = palavras[::-1] #inverti as palavras
texto_invertido = " ".join(palavras_invertidas) # juntei as palavras invertidas e concatenei com o espaço em braco
print(texto_invertido) # mostrando o resultado
```

```python
vogais = "aeiouAEIOU"
novo_texto = ""
for letra in a:
    if letra in vogais:
        novo_texto += "*"
    else:
        novo_texto += letra
print(novo_texto)
```

```python
palavras = a.split()
palavras_maiusculas = [palavra for palavra in palavras if palavra[0].isupper()]
texto_maiusculo = " ".join(palavras_maiusculas)
print(texto_maiusculo)
```
